```python
import numpy as np 
import pandas as pd 
```

```python
dataset = pd.read_csv('BigBasket.csv', header=None)
```

```python
transactions = []

for i in range(0, 7219):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
from apyori import apriori
market_basket = apriori(transactions=transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2, max_length=2)
```

```python
result = list(market_basket)
```

```python
result
```

```python
def inspect(result):
    product1         = [tuple(result[2][0][0])[0] for result in result]
    product2         = [tuple(result[2][0][1])[0] for result in result]
    supports    = [result[1] for result in result]
    confidences = [result[2][0][2] for result in result]
    lifts       = [result[2][0][3] for result in result]
    return list(zip(product1, product2, supports, confidences, lifts))
DataFrame_intelligence = pd.DataFrame(inspect(result), columns = ['product1', 'product1', 'Support', 'Confidence', 'Lift'])
```
