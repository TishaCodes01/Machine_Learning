```python
import pandas as pd 
```

```python
dataset = pd.read_csv('Amazon.com Clusturing Model.csv')
```

```python
X = dataset.iloc[:, [2,4]].values
```

```python
from sklearn.cluster import SpectralBiclustering
cluster = SpectralBiclustering(n_clusters=2, method='bistochastic', n_components=6, n_best=3, svd_method='randomized')
```

```python
cluster.fit(X)
```

```python
row_clusters = cluster.row_labels_
column_clusters = cluster.column_labels_
```

```python
X_reordered = X[row_clusters.argsort(), :][:, column_clusters.argsort()]
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_reordered, cmap='viridis', aspect='auto')
plt.title("Spectral Biclustering Result")
plt.xlabel("Features")
plt.ylabel("Samples")
plt.colorbar()
plt.show()
```
