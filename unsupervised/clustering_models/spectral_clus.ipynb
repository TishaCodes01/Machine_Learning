```python
import pandas as pd 
```

```python
dataset = pd.read_csv('Amazon.com Clusturing Model.csv')
```

```python
X = dataset.iloc[:, [2,4]].values
```

```python
from sklearn.cluster import SpectralClustering
cluster = SpectralClustering(n_clusters=5, affinity='nearest_neighbors', random_state=21)
cluster_fit = cluster.fit_predict(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=cluster_fit, cmap='viridis', s=50)
plt.title('Spectral Clustering')
plt.xlabel('Age')
plt.ylabel('Rating')
```
