```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('Amazon.com Clusturing Model.csv')
```

```python
X = dataset.iloc[:, 2:4].values
```

```python
from sklearn.cluster import FeatureAgglomeration
feature_agglomerator = FeatureAgglomeration(n_clusters=2)
X_transformed = feature_agglomerator.fit_transform(X)
```

```python
# Print the shape before and after
print(f"Original shape: {X.shape}")
print(f"Transformed shape: {X_transformed.shape}")
```

```python

plt.scatter(X_transformed[:, 0], X_transformed[:, 1], cmap='viridis')
plt.title("Feature Agglomeration Result")
plt.xlabel("Age")
plt.ylabel("Rating")
plt.show()
```
