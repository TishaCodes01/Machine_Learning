```python
import pandas as pd 
```

```python
dataset = pd.read_csv('Amazon.com Clusturing Model.csv')
```

```python
X = dataset.iloc[:, [2,4]].values
```

```python
from sklearn.cluster import SpectralCoclustering
cluster = SpectralCoclustering(n_clusters=3, random_state=21, svd_method='randomized', init='k-means++', n_init=10)
```

```python
cluster.fit(X)
```

```python
row_order = cluster.row_labels_
column_order = cluster.column_labels_
```

```python
reordered_data = X[row_order, :][:, column_order]
```

```python
import matplotlib.pyplot as plt
plt.imshow(reordered_data, aspect='auto', cmap='viridis')
plt.title('SpectralCoClustering')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```
