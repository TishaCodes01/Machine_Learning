```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Amazon.com Clusturing Model.csv')
```

```python
X = dataset.iloc[:, [2,4]].values
```

optimal number of cluster via elbow method

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=21)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow method to determine number of cluster')
plt.xlabel('number of cluster')
plt.ylabel('wcss value')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=21)
ymeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[ymeans == 0, 0], X[ymeans == 0, 1], s=50, color='red', label='cluster 1')
plt.scatter(X[ymeans == 1, 0], X[ymeans == 1, 1], s=50, color='blue', label='cluster 2')
plt.scatter(X[ymeans == 2, 0], X[ymeans == 2, 1], s=50, color='green', label='cluster 3')
plt.scatter(X[ymeans == 3, 0], X[ymeans == 3, 1], s=50, color='pink', label='cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, color='grey', label='centroids')
plt.title('K Means Clustering')
plt.xlabel('Age')
plt.ylabel('Rating')
plt.legend()
plt.show()
```
