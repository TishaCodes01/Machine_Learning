```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('Amazon.com Clusturing Model.csv')
```

```python
X = dataset.iloc[:, [2,4]].values
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=5, min_samples=3, metric='euclidean', algorithm='ball_tree')
yscan = dbscan.fit_predict(X)
```

```python
print(yscan)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50, c=yscan, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Age')
plt.ylabel('Rating')
plt.show()
```
