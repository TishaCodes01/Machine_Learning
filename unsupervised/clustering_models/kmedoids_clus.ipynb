```python
import pandas as pd
```

```python
dataset = pd.read_csv('Amazon.com Clusturing Model.csv')
```

```python
X = dataset.iloc[:, [2,4]].values
```

```python
print(X)
```

```python
from pyclustering.cluster.kmedoids import kmedoids
initial_medoids = [5,15,25,35]  # Indices of initial medoids

# Apply K-Medoids clustering
kmedoids_instance = kmedoids(X, initial_medoids)
kmedoids_instance.process()
```

```python
# Get the results
clusters = kmedoids_instance.get_clusters()
medoids = kmedoids_instance.get_medoids()
```

```python
import matplotlib.pyplot as plt 
# Visualize the clusters and the medoids
for cluster in clusters:
    plt.scatter(X[cluster, 0], X[cluster, 1])

# Highlight the medoids
for medoid in medoids:
    plt.scatter(X[medoid, 0], X[medoid, 1], color='grey', s=200, marker='X', label='Medoids')

plt.title("K-Medoids Clustering")
plt.legend()
plt.show()
```
