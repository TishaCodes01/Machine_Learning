```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('Amazon.com Clusturing Model.csv')
X = dataset.iloc[:, [2,4]].values
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage 
plt.figure(figsize=(12,4))
dendrogram = dendrogram(linkage(X, method = 'ward'))
plt.title('Dedrogram for choosing number of cluster')
plt.xlabel('Age')
plt.ylabel('Eucledian Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, metric = 'euclidean', linkage='ward')
yhc = hc.fit_predict(X)
```

```python
plt.scatter(X[yhc == 0, 0], X[yhc == 0, 1], s=50, c='red', label='cluster 1')
plt.scatter(X[yhc == 1, 0], X[yhc == 1, 1], s=50, c='orange', label='cluster 2')
plt.scatter(X[yhc == 2, 0], X[yhc == 2, 1], s=50, c='blue', label='cluster 3')
plt.scatter(X[yhc == 3, 0], X[yhc == 3, 1], s=50, c='pink', label='cluster 4')
plt.scatter(X[yhc == 4, 0], X[yhc == 4, 1], s=50, c='green', label='cluster 5')
plt.title('Hierarchical Clustering')
plt.xlabel('Age')
plt.ylabel('Rating')
plt.legend()
plt.show()
```
